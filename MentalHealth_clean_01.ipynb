```python
import pandas as pd

og_df=pd.read_csv("MentalHealth_dataset.csv")
og_df.head()

#to show how many missing values there are (null)
og_df.info()
og_df.isna().sum()

#creating a copy of the original file in order to produce a new file called df_clean
df_clean = og_df.copy()

#remove duplicates in order to remove bias
df_clean = df_clean.drop_duplicates()

#creating a mapping dictionary
continent_map = {
    "Australia" : "Oceania",
    "Belgium" : "Europe",
    "Bosnia and Herzegovina" : "Europe",
    "Brazil" : "South America",
    "Canada" : "North America",
    "Colombia" : "South America",
    "Costa Rica" : "South America",
    "Croatia" : "Europe",
    "Czech Republic" : "Europe",
    "Denmark" : "Europe",
    "Finland" : "Europe",
    "France" : "Europe",
    "Georgia" : "Europe",
    "Germany" : "Europe",
    "Greece" : "Europe",
    "India" : "Asia",
    "Ireland" : "Europe",
    "Israel" : "Asia",
    "Italy" : "Europe",
    "Mexico" : "South America",
    "Moldova" : "Europe",
    "Netherlands" : "Europe",
    "New Zealand" : "Oceania",
    "Nigeria" : "Africa",
    "Philippines" : "Asia",
    "Poland" : "Europe",
    "Portugal" : "Europe",
    "Russia" : "Europe",
    "Singapore" : "Asia",
    "South Africa" : "Africa",
    "Sweden" : "Europe",
    "Switzerland" : "Europe",
    "Thailand" : "Asia",
    "United Kingdom" : "Europe",
    "United States" : "North America"
}
df_clean["Continent"] = df_clean["Country"].map(continent_map)

#we have no numerical values, so only need to fill in missing categorical values:
categorical_cols = df_clean.select_dtypes(include=['object', 'category']).columns
#filling in the missing values using mode
df_clean[categorical_cols] = df_clean[categorical_cols].fillna(df_clean[categorical_cols].mode().iloc[0])

#removing the variables we're not interested in
df_clean = df_clean.drop(
    columns=[
        "Timestamp",
        "Days_Indoors",
        "Changes_Habits",
        "Mood_Swings",
        "Work_Interest",
        "Social_Weakness",
        "Coping_Struggles",
        "mental_health_interview"
    ], 
    errors="ignore")

#moving the 'Continent' column to be next to 'Country'
continent = df_clean["Continent"]
df_clean = df_clean.drop(columns=["Continent"])
df_clean.insert(df_clean.columns.get_loc("Country") + 1, "Continent", continent)

#to save cleaned dataset
df_clean.to_csv("C:/Users/PC/Documents/clean_MH_data_vs03.csv", index=False)
df_clean.head()
```
